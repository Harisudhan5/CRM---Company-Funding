```python
df2 = df[(df["Category"]!="Analytics and BI")&(df["Category"]!="Cloud and Software Dev Tools")&(df["Category"]!="IT Mgt")].index
dfs = df.drop(df2)
```

```python
size_of_company = dfs["Size fCompany"]
category = dfs["Category"]

data = {}
for size, cat in zip(size_of_company, category):
    key = f"{size}-{cat}"
    data[key] = data.get(key, 0) + 1
x_labels = list(data.keys())
y_values = list(data.values())

plt.figure(figsize=(15, 6))

plt.bar(x_labels, y_values)
plt.xlabel('Category or type')
plt.ylabel('Company count or frequency')
plt.title('Category Distribution with category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
